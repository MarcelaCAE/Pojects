## Attrition Rate by Job Level

```python
# creating an absolute and relative frequency table using crosstab
# key insights: The attrition rate on entry level is 63.61%, which is the highest among all job levels. (lack of grow)
# The attrition rate is 44.83%, which is lower than the entry-level attrition, but still significant.(lack of grow)
# The attrition rate is 20.08%, the lowest among all job levels.


absolute_frequency_table = pd.crosstab(clean_df['Job Level'], clean_df['Attrition'])
percentage_frequency_table = pd.crosstab(clean_df['Job Level'], clean_df['Attrition'], normalize='index') * 100
percentage_frequency_table = percentage_frequency_table.round(2)
combined_table = pd.concat([absolute_frequency_table, percentage_frequency_table], axis=1, keys=["Count", "Percentage"])
combined_table


```

```python
# pie chart attrition by job level 

import matplotlib.pyplot as plt

colors = ['#008080', '#66b2b2']
fig, axes = plt.subplots(1, len(absolute_frequency_table), figsize=(15, 7))
for i, job_level in enumerate(absolute_frequency_table.index):
    counts = absolute_frequency_table.loc[job_level]

    axes[i].pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops={'edgecolor': 'black'})
    axes[i].set_title(f'Attrition Distribution - {job_level}')

plt.tight_layout()
plt.show()

#similars colors - change color
```

```python
# avg years it's the same for each level
average_years_at_company = clean_df.groupby('Job Level')['Years at Company'].mean()
print(round(average_years_at_company))
```

# descriptive statitics - years at company

```python
# descriptive statitics - years at company

Q1 = clean_df['Years at Company'].quantile(0.25)
Q3 = clean_df['Years at Company'].quantile(0.75)
IQR = Q3 - Q1

#interquartile range - 50% centrality - measures the outliers
lower_limit  = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR


# outliers


upper_limit # 47
lower_limit # -17
outliers = clean_df[(clean_df['Years at Company'] < lower_limit) | (clean_df['Years at Company'] > upper_limit)]
num_outliers = outliers.shape[0]
num_outliers

```

Groppings

```python
# Adding the Job Code Colum based on the Job Role
job_role_counts = clean_df['Job Role'].value_counts()
clean_df['Job_Code'] = clean_df['Job Role'].map(job_role_counts.rank(method='dense').astype(int)) 
# metodo dense rank (rannks the values with more frequency starting by 1) map do the mapping of this ranking and astype convertes to a integer number
print(clean_df[['Job Role', 'Job_Code']].value_counts())

#Mapping by defining the ordinal data;
# Job Level
job_level_mapping = {'Entry': 1, 'Mid': 2, 'Senior': 3}
clean_df['Job_Level_Code'] = clean_df['Job Level'].map(job_level_mapping)

#work_life_balance
work_life_balance_mapping = {'Poor': 1, 'Fair': 2, 'Good': 3, 'Excellent': 4}
clean_df['Work_Life_Balance_Code'] = clean_df['Work-Life Balance'].map(work_life_balance_mapping)

#company_reputation
company_reputation_mapping = {'Poor': 1, 'Fair': 2, 'Good': 3, 'Excellent': 4}
clean_df['Company_Reputation_Code'] = clean_df['Company Reputation'].map(company_reputation_mapping)

#employee_recognition_mapping
employee_recognition_mapping = {'Low': 1, 'Medium': 2, 'High': 3, 'Very High': 4}
clean_df['Employee_Recognition_Code'] = clean_df['Employee Recognition'].map(employee_recognition_mapping)


# bins years of company
bins = [0,5,15,51] # pd.cut will creat the new groups of the years on a new column
labels = ['1-5', '6-15', '16+']
clean_df['Years_at_Company_Category'] = pd.cut(clean_df['Years at Company'], bins=bins, labels=labels)
clean_df['Years_at_Company_Category'].value_counts()

# Create a mapping for the 'Performance Rating' column
performance_rating_mapping = {'Low': 1, 'Below Average': 2, 'Average': 3, 'High': 4}
clean_df['Performance_Rating_Code'] = clean_df['Performance Rating'].map(performance_rating_mapping)
print(clean_df[['Performance Rating', 'Performance_Rating_Code']].head())

# Mapping for Gender column
gender_mapping = {'Male': 0, 'Female': 1, 'Other': 2}
clean_df['Gender_Code'] = clean_df['Gender'].map(gender_mapping)
print(clean_df[['Gender', 'Gender_Code']].head())

# Map for Attrition 
attrition_mapping = {'Stayed': 0, 'Left': 1}
clean_df['Attrition_Code'] = clean_df['Attrition'].map(attrition_mapping)
print(clean_df[['Attrition', 'Attrition_Code']].head())



```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.colors as mcolors


numerical_cols = clean_df.select_dtypes(include=['number']).columns
correlation_matrix = clean_df[numerical_cols].corr()
custom_palette = [
    (178/255, 216/255, 216/255),  # #b2d8d8
    (102/255, 178/255, 178/255),  # #66b2b2
    (0/255, 128/255, 128/255),    # #008080
    (0/255, 102/255, 102/255),    # #006666
    (0/255, 76/255, 76/255)       # #004c4c
]


cmap = mcolors.LinearSegmentedColormap.from_list("custom_cmap", custom_palette)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap=cmap, fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
